```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Exercise 1
### Filter

```python
img_ex1 = cv2.imread("Figuras/Fig0333a.tif", cv2.IMREAD_GRAYSCALE)
(h, w) = img_ex1.shape
```

```python
m = 25
d= int((m-1)/2)
kernel = np.ones((m,m),dtype ="float")/m**2

img_out =np.zeros((h,w),dtype = "uint8")
```

```python
for i in range(d,h-d):
    for j in range (d,w-d):
        secao_img = img_ex1[i-d:i+d+1,j-d:j+d+1]

        product =kernel*secao_img
        sum = product.sum()

        img_out[i,j]= sum.astype("uint8")
```

```python
plt.subplot(1,2,1)
plt.imshow(img_ex1, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_out, cmap='gray')
plt.title('Smooth Image')
plt.axis('off')

plt.show()
```

# Exercise 2
### Filter hubble

```python
img_ex2 = cv2.imread("Figuras/Fig0334a.tif", cv2.IMREAD_GRAYSCALE)
(h, w) = img_ex2.shape

m = 15
d= int((m-1)/2)
kernel = np.ones((m,m),dtype ="float")/m**2

img_filter=np.zeros((h,w),dtype = "uint8")

for i in range(d,h-d):
    for j in range (d,w-d):
        secao_img = img_ex2[i-d:i+d+1,j-d:j+d+1]

        product =kernel*secao_img
        sum = product.sum()

        img_filter[i,j]= sum.astype("uint8")


fig_bin =np.where(img_filter>img_filter.max()*0.35,255,0)

fig_out =np.where(fig_bin==255,img_ex2,0)
plt.figure(figsize=(15,10))
plt.subplot(1,4,1)
plt.imshow(img_ex2, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(img_filter, cmap='gray')
plt.title('Smooth')
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(fig_bin, cmap='gray')
plt.title('Binnary')
plt.axis('off')

plt.subplot(1,4,4)
plt.imshow(fig_out, cmap='gray')
plt.title('Final')
plt.axis('off')

plt.tight_layout()
plt.show()
```
