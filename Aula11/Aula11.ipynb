```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from library.selectBlob import selectBlob

img = cv2.imread("figuras/Alu.bmp", cv2.IMREAD_GRAYSCALE)

img_bin = np.where(img < 70, 255, 0).astype(np.uint8)

plt.figure(figsize=(15,10)) 
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Imagem Entrada')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_bin, cmap='gray')
plt.title('Imagem Saida')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Set up the detector with default parameters.
params = cv2.SimpleBlobDetector_Params()
# Set blob color (0=black, 255=white)
params.filterByColor = True
params.blobColor = 255
# Filter by Area
params.filterByArea = False
# params.minArea = 1
# #params.maxArea = 20000
# Filter by Circularity
params.filterByCircularity = False
# params.minCircularity = 0.8
# #params.maxCircularity = 1.2
# Filter by Convexity
params.filterByConvexity = False
#params.minConvexity = 0.87
#params.maxConvexity = 1
# Filter by Inertia
params.filterByInertia = False
#params.minInertiaRatio = 0.01
#params.maxInertiaRatio = 1
# Set up the detector with default parameters.
detector = cv2.SimpleBlobDetector_create(params)
# Detect blobs
KP = detector.detect(img_bin)
print("Nro de blobs: ",len(KP))
```

```python
# List parameters (X,Y,size,ang) of each detected keypoints
img1_text = cv2.cvtColor(img_bin,cv2.COLOR_GRAY2RGB)
i=1
for KPi in KP:
    print("Blob_", i, ": X= ", KPi.pt[0], " Y= ", KPi.pt[1], " size=", KPi.size**2, " ang=", KPi.angle)
    img1_text = cv2.putText(img1_text, str(i), (int(KPi.pt[0]),int(KPi.pt[1])), cv2.FONT_HERSHEY_PLAIN, 1, (0,0,255))
    i=i+1
#display image (with text)
plt.figure(figsize=(15,10)) 
plt.imshow(img1_text, cmap='gray')
plt.title('Imagem Entrada')
plt.axis('off')
plt.show()
#cv2.imshow("Img1 with texts", img1_text)
#aguarda uma tecla
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
# Draw detected blobs as red circles.
img1_with_KPs = cv2.drawKeypoints(img_bin, KP, np.array([]), (255,0,0), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.figure(figsize=(15,10)) 
plt.imshow(img1_with_KPs, cmap='gray')
plt.title('Imagem Entrada')
plt.axis('off')
plt.show()
```
