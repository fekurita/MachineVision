```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Exercise 1a

```python
img_ex1 = cv2.imread("Figuras_Aula6/Fig0342a.tif", cv2.IMREAD_GRAYSCALE)
(h, w) = img_ex1.shape

m = 3
d= int((m-1)/2)
kernel_Y = np.array([[-1,-1,-1],[0,0,0],[1,1,1]],dtype="int16")
kernel_X = np.array([[1,0,-1],[1,0,-1],[1,0,-1]],dtype="int16")

img_filter=np.zeros((h,w),dtype = "uint8")
img_filter2=np.zeros((h,w),dtype = "uint8")

for i in range(d,h-d):
    for j in range (d,w-d):
        secao_img = img_ex1[i-d:i+d+1,j-d:j+d+1]

        product =kernel_X*secao_img
        sum = product.sum()
        img_filter[i,j]= abs(sum).astype("uint8")
        
        product2 =kernel_Y*secao_img
        sum2 = product2.sum()
        img_filter2[i,j]= abs(sum2).astype("uint8")

img_junto = img_filter+img_filter2
plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.imshow(img_ex1, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_junto, cmap='gray')
plt.title('Smooth')
plt.axis('off')


plt.tight_layout()
plt.show()
```
