```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Exercersise 1
### Invert an image

```python
img_ex1 = cv2.imread("Figuras4/Fig0304a.tif", cv2.IMREAD_GRAYSCALE)
(h, w) = img_ex1.shape
```

```python
img_ex1_inv = np.zeros((h,w),dtype = "uint8")
```

```python
for i in range(h):
    for j in range (w):
        img_ex1_inv[i,j] = 255 -img_ex1[i,j]
```

```python
plt.subplot(1,2,1)
plt.imshow(img_ex1, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_ex1_inv, cmap='gray')
plt.title('Inverted Image')
plt.axis('off')

plt.show()
```

# Exercise 2
### Change Brightness and contrast of an image

```python
img_ex2 = cv2.imread("Figuras4/Fig2.tif", cv2.IMREAD_GRAYSCALE)
(h, w) = img_ex2.shape
```

```python
mult =3
soma =30
```

```python
img_ex2_32 = img_ex2.astype(np.int32)
img_ex2_out = np.zeros((h,w),dtype = "uint8")
```

```python
for i in range(h):
    for j in range(w):
        intens32 = img_ex2_32[i,j]*mult+soma
        img_ex2_out[i,j]=np.clip(intens32,0,255).astype(np.uint8)
```

```python
plt.subplot(1,2,1)
plt.imshow(img_ex2, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_ex2_out, cmap='gray')
plt.title('Bright')
plt.axis('off')

plt.show()
```

# Exercise 3a
### Power-law transformation (gama<1)

```python
img_ex3a = cv2.imread("Figuras4/Fig0305a.tif", cv2.IMREAD_GRAYSCALE)
(h, w) = img_ex3a.shape
```

```python
gama=0.1
```

```python
img_ex3a_float = img_ex3a.astype(np.float)
img_ex3a_float = img_ex3a_float/255
img_ex3a_out = np.zeros((h,w),dtype = "uint8")
```

```python
for i in range(h):
    for j in range(w):
        img_ex3a_float[i,j] = img_ex3a_float[i,j]**gama
        img_ex3a_float[i,j] = img_ex3a_float[i,j]*255
        img_ex3a_out[i,j]=img_ex3a_float[i,j].astype(np.uint8)
```

```python
plt.subplot(1,2,1)
plt.imshow(img_ex3a, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_ex3a_out, cmap='gray')
plt.title('Intensefied')
plt.axis('off')

plt.show()
```

# Exercise 4
### Image substraction

```python
img_ex4A = cv2.imread("Figuras4/Fig4A.png", cv2.IMREAD_GRAYSCALE)
img_ex4B = cv2.imread("Figuras4/Fig4B.png", cv2.IMREAD_GRAYSCALE)
(h, w) = img_ex4A.shape
img_ex4_out = np.zeros((h,w),dtype = "uint8")
for i in range(h):
    for j in range(w):
        img_ex4_out[i,j]=img_ex4A[i,j]-img_ex4B[i,j]

plt.subplot(1,3,1)
plt.imshow(img_ex4A, cmap='gray')
plt.title('Original Image 1')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(img_ex4B, cmap='gray')
plt.title('Original Image 2')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(img_ex4_out, cmap='gray')
plt.title('Image output')
plt.axis('off') 
plt.show()
        
```
